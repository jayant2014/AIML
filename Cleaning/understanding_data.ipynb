# Normalizing Data

```python
height = data.height
```

```python
norm = (height - height.mean()) / height.std()
```

```python
norm
```

```python
minmax = (height - height.min()) / (height.max() - height.min())
```

```python
minmax
```

```python
minmax.min()
```

```python
minmax.max()
```

```python
data.height = minmax
```

```python
data['standardized_height'] = norm
```

```python
data.newcolumn = norm
```
